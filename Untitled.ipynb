```python
import torch
import matplotlib.pyplot as plt
```

```python
model = torch.nn.Linear(100, 200)
optimizer = torch.optim.Adam(model.parameters(), lr=0.025)

data_size, batch_size, num_epochs = 200_000, 64, 5
num_steps = int(data_size / batch_size) * num_epochs
scheduler = torch.optim.lr_scheduler.CyclicLR(optimizer, 0.0001, 0.001, int(num_steps * 0.1),
                                              mode="exp_range", gamma=0.9999)
```

```python
lrs = [scheduler.get_last_lr()]
for _ in range(num_steps):
    scheduler.step()
    lrs.append(scheduler.get_last_lr())
```

```python
plt.plot(lrs, c="b")
plt.xlabel("Training step (in batches)")
plt.ylabel("Learning rate")
plt.title("Learning rate with Cyclic scheduler")
plt.grid()
```
